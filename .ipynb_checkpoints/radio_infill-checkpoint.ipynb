```python
#Thank u simon
#Thank u simon
def yc_from_m_z(M500c,z):
    tenToA0=4.95e-5
    B0=0.08
    Mpivot=3
    #Q=calcQ(theta500cArcmin, med_tckQFit)
    Ez=cosmo.efunc(z)
    fixed_y_c=tenToA0*np.power(Ez, 2)*np.power(M500c/Mpivot, 1+B0)*Q(M500c, z)
    return fixed_y_c

data = np.load('Q_grid_tile1_10_8.npz')

zs = data['z']    
Qarr = data['Qarr']    
mass = data['mass']    


Qarr = Qarr[1:,0:]
Y = mass[0:]
X = zs[1:]

#x,y = np.meshgrid(X,Y)

#Q = interpolate.interp2d(x,y,Qarr,kind='cubic')

#Xnew, Ynew = np.linspace(min(zs), max(zs), 100), np.linspace(min(mass), max(mass), 100)

def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return idx


def Q(M500c, z):
    m_idx = find_nearest(mass, M500c)
    z_idx = find_nearest(zs, z)
    return(Qarr[z_idx, m_idx])

vec_yc_from_m_z = np.vectorize(yc_from_m_z)
```

```python

def get_nearby_radio(cluster_coords, radio_coords, r = 2.5):
    r = r*np.pi/(60*180)
    
    ra, dec = cluster_coords
    
    dr = np.sqrt((ra-radio_coords[0])**2 + (dec-radio_coords[1])**2)
    
    flags = np.where((dr <= r))[0]
        
    return flags, dr[flags]
```

```python
#Load halo data
from   cosmology import *

rho = 2.775e11*omegam*h**2 # Msun/Mpc^3
path = '/scratch/r/rbond/msyriac/data/sims/websky/v0.0/'

f=open(path + 'halos.pksc')
N=np.fromfile(f,count=3,dtype=np.int32)[0]

# only take first five entries for testing (there are ~8e8 halos total...)
# comment the following line to read in all halos
N = 50000

catalog=np.fromfile(f,count=N*10,dtype=np.float32)
catalog=np.reshape(catalog,(N,10))

x  = catalog[:,0];  y = catalog[:,1];  z = catalog[:,2] # Mpc (comoving)
vx = catalog[:,3]; vy = catalog[:,4]; vz = catalog[:,5] # km/sec
R  = catalog[:,6] # Mpc

# convert to mass, comoving distance, radial velocity, redshfit, RA and DEc
M        = 4*np.pi/3.*rho*R**3        # Msun
chi      = np.sqrt(x**2+y**2+z**2)    # Mpc
vrad     = (x*vx + y*vy + z*vz) / chi # km/sec
redshift = zofchi(chi)
#DEC, RA  = hp.vec2ang(np.column_stack((x,y,z)),lonlat=True)
yc       = vec_yc_from_m_z(M/1e14, redshift)

# convert to ra / dec (radians) from readhalos.py
colat, cluster_ras = hp.vec2ang(np.column_stack((x, y, z)))

# convert colat to dec 
cluster_decs = np.pi/2 - colat

```

```python
cluster_cut = 3.8e14 #Hilton21 90% completeness

cluster_flags = np.where((M >= cluster_cut))[0]

M            = M[cluster_flags]
redshift     = redshift[cluster_flags]
cluster_decs = cluster_decs[cluster_flags]
cluster_ras  = cluster_ras[cluster_flags]
yc           = yc[cluster_flags]
```

```python
r_cut = 5

drs = []

all_fluxes = []
for i in range(len(cluster_ras)):
    flags, dr = get_nearby_radio([cluster_ras[i], cluster_decs[i]], [radio_ras, radio_decs], r = r_cut)
    if len(dr) == 0:
        drs.append([0])
        all_fluxes.append([0])
    else:
        drs.append(np.array(dr))
        all_fluxes.append(np.array(fluxes[flags]))
    
drs = np.array(drs, dtype=object)
all_fluxes = np.array(all_fluxes, dtype=object)
```

```python
max_fluxes = np.zeros(len(all_fluxes))
total_fluxes = np.zeros(len(all_fluxes))

for i in range(len(all_fluxes)):
    max_fluxes[i] = np.max(all_fluxes[i])
    total_fluxes[i] = sum(all_fluxes[i])
```

```python
from astropy.table import QTable
import pylab as pl
pl.ion()
import numpy as np
from scipy import interpolate
from scipy import stats

diag_plot = False

#first plot those figures: 

wkDir='/home/r/rbond/jorlo/dev/websky_model/'

####read in data files........
talpha=QTable.read(wkDir+'sim_fits/alphas_1mJy_central_source.fits')
tSync=QTable.read(wkDir+'sim_fits/central_source_alpha=-0.7.fits')
tDust=QTable.read(wkDir+'sim_fits/central_source_alpha=3.5.fits')
tOffset=QTable.read(wkDir+'sim_fits/offset_1mJy_alpha=-0.7.fits')
tOffset2=QTable.read(wkDir+'sim_fits/offset_1mJy_alpha=3.5.fits')
o0p1=QTable.read('sim_fits/offset_0.2-mJy_alpha=0.1.fits')
o0p7=QTable.read('sim_fits/offset_0.2-mJy_alpha=-0.7.fits')
o3p5=QTable.read('sim_fits/offset_0.2-mJy_alpha=3.5.fits')
talpha=QTable.read(wkDir+'sim_fits/alphas_1mJy_central_source.fits')
```

```python
#############################################################################
#Bin simulations

d3p5=np.unique(o3p5['offset_arcsec'])
r3p5=np.zeros(len(d3p5))
i3p5=np.zeros(len(d3p5))
for i in range(len(d3p5)):
    p=np.where((o3p5['offset_arcsec'] == d3p5[i]))
    r3p5[i]=np.median(o3p5['srcinj_fixed_y_c'][p]-o3p5['fixed_y_c'][p])
    i3p5[i]=stats.iqr((o3p5['srcinj_fixed_y_c'][p]-o3p5['fixed_y_c'][p]),rng=(16.,84.)) / 2.
i3p5n = i3p5 / np.abs(np.min(r3p5))
r3p5n = r3p5 / np.abs(np.min(r3p5))

d0p7=np.unique(o0p7['offset_arcsec'])
r0p7=np.zeros(len(d0p7))
i0p7=np.zeros(len(d0p7))
for i in range(len(d0p7)):
    p=np.where((o0p7['offset_arcsec'] == d0p7[i]))
    r0p7[i]=np.median(o0p7['srcinj_fixed_y_c'][p]-o0p7['fixed_y_c'][p])
    i0p7[i]=stats.iqr((o0p7['srcinj_fixed_y_c'][p]-o0p7['fixed_y_c'][p]),rng=(16.,84.)) / 2.
i0p7n = i0p7 / np.abs(np.min(r0p7))
r0p7n = r0p7 / np.abs(np.min(r0p7))

d0p1=np.unique(o0p1['offset_arcsec'])
r0p1=np.zeros(len(d0p1))
i0p1=np.zeros(len(d0p1))
for i in range(len(d0p1)):
    p=np.where((o0p1['offset_arcsec'] == d0p1[i]))
    r0p1[i]=np.median(o0p1['srcinj_fixed_y_c'][p]-o0p1['fixed_y_c'][p])
    i0p1[i]=stats.iqr((o0p1['srcinj_fixed_y_c'][p]-o0p1['fixed_y_c'][p]),rng=(16.,84.)) / 2.
i0p1n = i0p1 / np.abs(np.min(r0p1))
r0p1n = r0p1 / np.abs(np.min(r0p1))
```

```python
#fit to Mats simulations
#note very similar results for all spectral indexes *assuming* delta y_c does not get too large.
N_r=interpolate.CubicSpline(d0p7,-r0p7n)
dr=np.arange(-300,310,10)
diag_plot = False
if diag_plot:
    fig=pl.figure(figsize=(4.27,3.2))#pl.figure()
    #pl.errorbar(offs,respn,yerr=iqrn,label='alpha=-0.7',color='b',marker='o')
    #pl.errorbar(offs3p5,respn3p5,yerr=iqrn3p5,label='alpha=3.5',color='r',marker='o')
    pl.errorbar(d0p7,-r0p7n,yerr=i0p7n,label='alpha=-0.7',color='k',marker='o',linestyle='')
    pl.plot(dr,N_r(dr),c='k')
    #pl.errorbar(d3p5,r3p5n,yerr=i3p5n,label='alpha=3.5',color='r',marker='o')
    #pl.errorbar(d0p1,r0p1n,yerr=i0p1n,label='alpha=0.1',color='g',marker='o')
    #pl.legend()
    pl.title(r'Effects of decentering of sources',fontsize='medium')
    pl.xlabel(r'Distance from cluster center / arcsec')
    pl.ylabel(r'Normalized response : $N(r)$')
    fig.tight_layout()
```

```python
####alphas####
diag_plot = False
talpha=QTable.read(wkDir+'sim_fits/alphas_1mJy_central_source.fits')

delta0=-0.08756e-4 #reference everything to 1mJy at alpha=0.7

alpha=np.unique(talpha['alpha'])
A=np.zeros(len(alpha))
Aiqr=np.zeros(len(alpha))
for i in range(len(alpha)):
    p=np.where((talpha['alpha'] == alpha[i]))
    A[i]=np.median(talpha['srcinj_fixed_y_c'][p]-talpha['fixed_y_c'][p])
    Aiqr[i]=stats.iqr((talpha['srcinj_fixed_y_c'][p]-talpha['fixed_y_c'][p]),rng=(16.,84.)) / 2.
A_alpha=interpolate.CubicSpline(alpha,A/delta0)
x=np.arange(-1.5,4.1,0.1)
if diag_plot:
    fig=pl.figure(figsize=(4.27,3.2))#pl.figure()
    pl.errorbar(alpha,A/delta0,yerr=Aiqr/delta0,marker='o',linestyle='',color='b')
    pl.plot(x,A_alpha(x),c='b')
    pl.title(r'$\Delta \tilde{y}_0$ with spectral index',fontsize='medium')
    pl.xlabel(r'spectral index ($\alpha$)')
    pl.ylabel(r'Normalized $\Delta \tilde{y}_0$ :  $A(\alpha)$')
    pl.text(-1,5,r'$A(\alpha) = 1.0$ for $\alpha=-0.7$')
    fig.tight_layout()
```
